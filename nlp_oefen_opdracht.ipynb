```python
import pandas as pd

with open("data/alice.txt") as f:
    engels = f.readlines()
with open("data/zwendelaars.txt") as f:
    nederlands = f.readlines()


def create_bigrammen(tekst):
    pair = {}
    for zin in tekst:
        zin = " " + zin.lower() + " "
        zin = zin.translate(str.maketrans("", "", "().,!?'˜-"))
        for i in range(len(zin)-1):
            wordpair = zin[i] + zin[i+1]
            if wordpair in pair:
                pair[wordpair] += 1
            else:
                pair[wordpair] = 1
    return pair
```

```python
from collections import Counter



def create_monogrammen(tekst):
    mono_dict = {}

    for zin in tekst:
        zin = zin.lower()
        zin = zin.translate(str.maketrans("", "", "().,!?'˜-"))

        for char in zin:
            mono_dict[char] = mono_dict.get(char, 0) + 1

    return mono_dict
```

```python
engels_pairs = create_bigrammen(engels)
nederlands_pairs = create_bigrammen(nederlands)

engels_mono = create_monogrammen(engels)
```
